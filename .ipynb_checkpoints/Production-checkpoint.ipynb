### Import Dependancies 

```python
import pandas as pd
import os
import numpy as np
```

```python
# Import Data 
production = "Resources/Production.csv"

# Read data into Pandas DataFrames
production_df = pd.read_csv(production)
production_df.head()
```

### Clean Data

```python
# Drop Null Values 
production_df = production_df.dropna()
```

```python
# Round Volume to nearest 100
production_df['Production Volume'] = (production_df['Production Volume'] / 100).round() * 100
production_df.head()
```

```python
# Calculate Total Production cost

production_df['Total Cost'] = ((production_df['Direct Material'] + 
                               production_df['Direct Labour'] + 
                               production_df['Variable Overhead']) * 
                               production_df['Production Volume']) + production_df['Fixed Overhead']
production_df.head()
```

```python
#Calculate Cost per unit

production_df['Cost Per Unit'] = (production_df['Total Cost'] / production_df['Production Volume']).round(2)

production_df
```

```python
production_df.to_csv('Product Cost Per Unit.csv', index=False)
```

### Clustering Based on Cost Per Unit

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt

# Sample DataFrame
# df = pd.DataFrame(...)  # this is just a placeholder, you already have production_df

# One-hot encoding the 'Product' column
product_encoded = pd.get_dummies(production_df['Product'], prefix='Product')

# Extract necessary columns and concatenate the encoded product data
X = pd.concat([production_df[['Production Volume', 'Cost Per Unit']], product_encoded], axis=1)

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Choose the number of clusters (for this example, let's use 3)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
production_df['Cluster'] = kmeans.fit_predict(X_scaled)

# Plotting the clusters for visualization
for cluster_num in range(n_clusters):
    cluster_data = production_df[production_df['Cluster'] == cluster_num]
    plt.scatter(cluster_data['Production Volume'], cluster_data['Cost Per Unit'], label=f'Cluster {cluster_num}')

plt.xlabel('Production Volume')
plt.ylabel('Cost Per Unit')
plt.title('KMeans Clustering based on Cost Per Unit and Production Volume')
plt.legend()
plt.show()
```

```python
print(production_df[['Product', 'Cluster']])
```

```python
clustered_products_df = production_df[['Product', 'Cluster']]
clustered_products_df
```
