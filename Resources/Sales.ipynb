### Import Dependancies

```python
import pandas as pd
import os
import numpy as np
```

```python
# Import Data 
sales = "Sales.csv"

# Read data into Pandas DataFrames
sales_df = pd.read_csv(sales)
sales_df.head()
```

### Clean DataÂ¶

```python
#Round Null Values 
sales_df = sales_df.dropna()
sales_df.head()
```

```python
# Round Volume to nearest 100
sales_df['Total'] = (sales_df['Total'] / 100).round() * 100
sales_df.head()
```

```python
sales_df.to_csv('../Cleaned_Data/Sales.csv', index=False)
```
